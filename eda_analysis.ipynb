**Exploratory Data Analysis (EDA)** is an approach to analyzing and summarizing data sets in order to gain insights into the data. Here are 11 steps to perform EDA:

1. **Understand the problem and the data set.**

2. **Import necessary libraries and load the data set.**

3. **Check the data set's shape and size.**

4. **Check for missing values and handle them.**

5. **Check for duplicates and handle them.**

6. **Check for outliers and handle them.**

7. **Check for data types and convert them if necessary.**

8. **Check for correlations between variables.**

9. **Visualize the data set using various plots and charts.**

10. **Draw conclusions and make recommendations based on the analysis.**

11. **Share the results.**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # plotting
import seaborn as sns  # more data visualization tools
import warnings # warning filter
warnings.filterwarnings('ignore') # warnings are ignored
df = sns.load_dataset('titanic') # load the data set

print('Shape:', df.shape) # Step 1: Get to know the data set
print('Size:', df.size)  # Shape: (891, 15)

print('Missing values:', df.isnull().sum()) # Step 2: Check for missing values

print('Duplicates:', df.duplicated().sum()) # Step 3: Check for duplicates

df = df.drop_duplicates() # Drop duplicates
print('Duplicates:', df.duplicated().sum()) # Check for duplicates
sns.boxplot(x=df['age'], y=df['survived'], orient='h', palette='Set2') # Step 4: Plot the data

df['sex']=df['sex'].astype('category').cat.codes  # Step 5: Convert categorical variables to numeric
df['embarked']=df['embarked'].astype('category').cat.codes # Convert categorical variables to numeric

corr_matrix = df.corr() # Step 6: Compute the correlation matrix
print(corr_matrix) # Print the correlation matrix

print('Strong correlations:', corr_matrix['survived'].abs().sort_values(ascending=False).head(4)) # Print the top 4 correlations
print('Important predictors:', corr_matrix['survived'].abs().sort_values(ascending=False).head(4).index.tolist()) # Convert the index to a list

# head of all columns 
for col in df.columns:
    print(df[col].head()) # Step 7: Check for outliers
    print('\n')
df.to_csv('titanic_clean.csv') # Save the cleaned data set

plt.barh(df['embarked'].unique(), df['embarked'].value_counts(), color='lime', edgecolor='black', alpha=0.5) # Plot the data
plt.xlabel('Frequency') # Add a label to the x-axis
plt.ylabel('Category') # Add a label to the y-axis
plt.title('Frequency of each category') # Add a title to the plot
plt.show() # Show the plot
sns.catplot(x='pclass', y='survived', hue='sex', kind='bar', data=df, palette='Set2') # Plot the data
plt.show() # Show the plot

sns.heatmap(corr_matrix, annot=True, cmap='autumn_r', linewidths=0.5, annot_kws={'size': 10}) # Plot the data
plt.xticks(rotation=45) # Rotate the x-axis labels
plt.yticks(rotation=45) # Rotate the y-axis labels
plt.show() # Show the plot
```

There are many common visualizations used in EDA. Here are some examples:

- Histograms: used to show the distribution of a numerical variable.
- Box plots: used to show the distribution of a numerical variable and identify outliers.
- Scatter plots: used to show the relationship between two numerical variables.
- Line plots: used to show the trend of a numerical variable over time.
- Bar plots: used to show the distribution of a categorical variable.
- Heat maps: used to show the correlation between variables.
- Pair plots: used to show the relationship between multiple variables.

Note that these are just some of the many visualizations that can be used in EDA, and the choice of visualization depends on the data set and the problem at hand.



## Parameters in Seaborn Heatmap

The following are the most commonly used parameters for creating a heatmap in Seaborn:

- `data`: The data set to be plotted as a heatmap.
- `cmap`: The color map to be used for the heatmap. The default is `None`, which uses the `viridis` color map.
- `center`: The value at which to center the color map. The default is `None`, which centers the color map at 0.
- `annot`: Whether to annotate the heatmap with the correlation coefficients. The default is `None`, which does not annotate the heatmap.
- `fmt`: The format string for the annotation text. The default is `'.2g'`, which formats the text as a floating point number with 2 decimal places.
- `linewidths`: The width of the lines separating the cells in the heatmap. The default is `0`, which does not draw any lines.
- `linecolor`: The color of the lines separating the cells in the heatmap. The default is `'white'`.
- `cbar`: Whether to draw a color bar next to the heatmap. The default is `True`.
- `cbar_kws`: Additional keyword arguments to be passed to the color bar.
- `square`: Whether to make the cells in the heatmap square. The default is `False`.
- `xticklabels`: Whether to show the x-axis tick labels. The default is `'auto'`, which shows the tick labels if there is enough space.
- `yticklabels`: Whether to show the y-axis tick labels. The default is `'auto'`, which shows the tick labels if there is enough space.
- `mask`: A boolean mask indicating which cells to mask out of the heatmap. The default is `None`, which does not mask any cells.
- `ax`: The matplotlib axis on which to draw the heatmap. The default is `None`, which creates a new axis.

## Values

The values in a heatmap represent the correlation coefficients between different variables in a data set. The correlation coefficient is a measure of the strength and direction of the linear relationship between two variables. The correlation coefficient can range from -1 to 1, with -1 indicating a perfect negative correlation, 0 indicating no correlation, and 1 indicating a perfect positive correlation.

In a heatmap, the correlation coefficients are color-coded according to a color map. The default color map in Seaborn is `viridis`, which ranges from dark blue for negative values to yellow for positive values. Other color maps that can be used include `coolwarm`, `magma`, `inferno`, and `plasma`, among others.
